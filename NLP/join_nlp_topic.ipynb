```python
import pandas as pd

# Cargar los CSVs
philly = pd.read_csv("philly.csv")
business = pd.read_csv("business.csv")

# Hacemos un inner join para quedarnos solo con los registros que existen en ambos datasets
merged = pd.merge(philly, business, on="business_id", how="inner")

merged.rename(columns={'stars': 'restaurant_stars'}, inplace=True)

# Orden de columnas deseado
desired_column_order = [
    'business_id', 'name', 'address', 'city', 'state', 'postal_code',
    'latitude', 'longitude', 'restaurant_stars', 'review_count', 'is_open',
    'attributes', 'categories', 'hours', 'review_id', 'user_id', 'useful',
    'funny', 'cool', 'text', 'date',
    'topic_0', 'topic_1', 'topic_2', 'topic_3', 'topic_4', 'topic_5'
]

# Añadir columnas faltantes como vacías (NaN)
for col in desired_column_order:
    if col not in merged.columns:
        merged[col] = pd.NA  # o np.nan si prefieres float-compatible

# Reordenar las columnas según el orden deseado
df = merged[desired_column_order]
len(df.columns)
```
